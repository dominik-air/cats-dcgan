```python
import tensorflow as tf

cat_generator = tf.keras.models.load_model('tf-hvd-cats-generator')
cat_generator.summary()
```

```python
import matplotlib.pyplot as plt
noise = tf.random.normal((16, 100))
img = cat_generator(noise)
print(img.shape)

fig, axes = plt.subplots(4, 4)

axes = axes.flatten()

fig.set_size_inches(8, 8)
for idx, ax in enumerate(axes):
    ax.imshow(img[idx])
    ax.axis('off')
plt.show()
```
